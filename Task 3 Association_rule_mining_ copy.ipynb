```python
as_df
```

```python
# converting the data frame with factors to one hot encoded format
as_onehot = pd.get_dummies(as_df)
```

```python
# one hot encoded data frame
as_onehot
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules

# writing the rules with support
frequent_item = apriori(as_onehot, min_support=0.02, use_colnames=True)

# generating the association rules with confidance 
rules = association_rules(frequent_item, metric = 'confidence', min_threshold=0.3)

```

```python
rules
```

```python
# sorting the rules based on highest confidance
rules = rules.sort_values(['confidence', 'lift', 'support'], ascending =[False, False, False])
```

```python
rules
```

```python
# printing the top rules with a minimum length of ascendents as 2 and consequent as the int_overall_rating_high.
rules['antecedent_len'] = rules['antecedents'].apply(lambda x: len(x))

# Display the association rules with required attributes.
rules_results = rules[(rules['antecedent_len'] >= 2) & 
                (rules['consequents'] == {'int_overall_rating_high'})][['antecedents', 'consequents', 'support', 'confidence', 'lift']]

rules_results
```
